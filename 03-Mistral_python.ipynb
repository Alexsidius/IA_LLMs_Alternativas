### Usando servi√ßos da Mistral

```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
import os

api_key = os.environ.get("MISTRAL_API_KEY")

from mistralai import Mistral

client = Mistral(api_key=api_key)

response = client.chat.complete(
    model="mistral-large-latest",
    messages=[
        {"role": "user", "content": "Explique para uma crian√ßa quem √© jesus cristo"}
    ],
)

print(response.choices[0].message.content)
```

### Vision
Usando leitura de imagem com o rosto de varias mulheres de diferentes estilo e culturas

```python
model = "mistral-small-latest"

client = Mistral(api_key=api_key)

messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": "descreva para mim o que esta imagem transmite"
            },
            {
                "type": "image_url",
                "image_url": "https://raw.githubusercontent.com/Alexsidius/IA_LLMs_Alternativas/refs/heads/master/files/imagemWM.png"
            }
        ]
    }
]

chat_response = client.chat.complete(
    model=model,
    messages=messages
)
```

```python
print(chat_response.choices[0].message.content) 
```
