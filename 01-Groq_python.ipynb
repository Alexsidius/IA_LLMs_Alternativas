### Demonstração de 3 serviços de IA usando GROQ
##### Geração de texto, descrição de audio e identificação de imagem

```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
```

```python
from groq import Groq
client = Groq()#api_key=os.getenv("GROQ_API_KEY"))
```

```python
menssagens = [
    {
        "role": "user",
        "content": "Explique a teoria da relatividade de forma simples."

    }
]
response = client.chat.completions.create(
    messages=menssagens,
    #llama-3.1-8b-instant — modelo leve e rápida, boa para teste ou uso com menos custo
    model="llama-3.3-70b-versatile",)

response
```

```python
#imprime a resposta apenas mensagem principal sem os parametros
print(response.choices[0].message.content)
```

### Stream
##### Abaixo maneira de gerar a resposta gradativamente, para respostas maiores e mais elaboradas.

```python
menssagens = [
    {
        "role": "user",
        "content": "Explique o paradoxo de bootstrap de forma simples."

    }
]
stream = client.chat.completions.create(
    messages=menssagens,
    #llama-3.1-8b-instant — modelo leve e rápida, boa para teste ou uso com menos custo
    model="llama-3.3-70b-versatile",
    stream=True,)

for chunck in stream:
    print(chunck.choices[0].delta.content, end="", flush=True)  
```

### Transcrição de audio

```python
#formatar a saida do texto para não prolongar para as laterais
import textwrap

def formatar_texto(response):
    text = response.text
    text_formatted = textwrap.fill(text, width=100)
    print(text_formatted)
        
```

```python
from groq import Groq

client = Groq() 

arquivo = "files/versiculo.mp4"

with open(arquivo, "rb") as audio:
    transcricao = client.audio.transcriptions.create(
        file=(arquivo, audio.read()),
        model="whisper-large-v3",
        response_format="json",
        language="pt",
        prompt = "esta é uma citação de um belo versículo biblico no livro de romanos.",
    )

    formatar_texto(transcricao)
```

### Vision - visão computacional

```python
#formatar a saida do texto para não prolongar para as laterais
import textwrap

def formatar_texto(response):
    text = response.content
    text_formatted = textwrap.fill(text, width=100)
    print(text_formatted)
        
```

```python
completion = client.chat.completions.create(
    model="llama-3.2-11b-vision-instruct",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {
                        "url": "https://raw.githubusercontent.com/Alexsidius/IA_LLMs_Alternativas/master/files/imagemCD.png"
                    }
                },
                {
                    "type": "text",
                    "text": "O que há nesta imagem?"
                }
            ]
        }
    ],
    max_tokens=500
)

print(completion.choices[0].message["content"])
```
