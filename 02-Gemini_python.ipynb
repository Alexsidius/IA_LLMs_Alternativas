### Leitura de imagens


```python
import httpx
import os
import base64
import google.generativeai as genai

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

model = genai.GenerativeModel("models/gemini-2.5-flash")

# URL RAW do GitHub (correta!)
image_url = "https://raw.githubusercontent.com/Alexsidius/IA_LLMs_Alternativas/master/files/imagemCD.png"

# Baixa a imagem
image = httpx.get(image_url).content

# Converte para Base64
image_b64 = base64.b64encode(image).decode("utf-8")

prompt = "Descreva a imagem."

# Envia apenas UMA parte de imagem + o prompt
response = model.generate_content([
    {"mime_type": "image/png", "data": image_b64},
    {"text": prompt}
])

print(response.text)
```
